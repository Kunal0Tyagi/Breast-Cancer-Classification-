```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

```

```python
brc_dataset = sklearn.datasets.load_breast_cancer()#loading the dataset from sklearn
print(brc_dataset)#priting the dataset in its raw state without dataframes
```

```python
df= pd.DataFrame(brc_dataset.data, columns = brc_dataset.feature_names)#loading the data in a dataframe 
df.head(10)#printing first 10 lines of the dataframe
```

```python
correlationmatrix=df.corr()
sns.heatmap(correlationmatrix,annot=False)
plt.show()
```

```python
df['label'] = brc_dataset.target#adding target(col) as label
df.tail()#prints the last 10 rows of the dataframe
```

```python
df.shape#for no of rows and cols in the datasets
df.isnull().sum()#check for any missing values in the dataset
```

```python
df.describe()
```

```python
# checking the distribution of Target Varibale with a value and a bar graph
print(df['label'].value_counts())
sns.countplot(df['label'], label = 'count')
print("\n1->Benign\n0->Malignant")
```

```python
sns.pairplot(df,hue = 'label', palette= 'coolwarm', vars = ['mean radius', 'mean texture', 'mean perimeter','mean area','mean smoothness'])
print("\n1->Benign\n0->Malignant")
```

```python
df.groupby('label').mean()
```

```python
X = df.drop(columns='label', axis=1)
Y = df['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)#splitting of data into training and test
print(X.shape, X_train.shape, X_test.shape)
```

```python

model = LogisticRegression()
# training the Logistic Regression model using Training data

model.fit(X_train, Y_train)
# accuracy on training data
X_train_pred = model.predict(X_train)
training_data_acc = accuracy_score(Y_train, X_train_pred)
print('Accuracy on training data(Logistic Regression) = ', training_data_acc)
# accuracy on test data
X_test_pred = model.predict(X_test)
test_data_acc = accuracy_score(Y_test, X_test_pred)
print('Accuracy on test data(Logistic Regression)= ', test_data_acc)
model1 = DecisionTreeClassifier()
model1.fit(X_train, Y_train)
X_train_pred = model1.predict(X_train)
training_data_acc = accuracy_score(Y_train, X_train_pred)
print('Accuracy on training data(Decision Tree Classifier) = ', training_data_acc)
# accuracy on test data
X_test_pred = model1.predict(X_test)
test_data_acc = accuracy_score(Y_test, X_test_pred)
print('Accuracy on test data(Decision Tree Classifier) = ', test_data_acc)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
y_pred=model.predict(X_test)
cm=confusion_matrix(Y_test,y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
print("Confusion matrix to check the accuracy on logistic regression model:")
disp.plot()
```
